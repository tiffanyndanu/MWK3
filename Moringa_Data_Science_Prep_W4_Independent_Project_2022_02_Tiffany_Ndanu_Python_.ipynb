## **DATA PREPARATION**

**IMPORTING LIBRARIES.**

```python
#We will first import our dataset that we will use for the examples. We will do this by first
# Importing our pandas library 

import pandas as pd
```

**LOADING DATASET.**

```python
#storing the dataset location to the variable url.
url =  'http://bit.ly/autolib_dataset'

#reading the url and storing the datframe in the variable df.
df = pd.read_csv(url)

#preview of the first five rows
df.head(5)
```

# **DATA CLEANING.**

```python
#Dropping unnecessary columns.

df_clean = df.drop(columns = ['Cars' , 'Displayed comment' ,	'ID' ,'Geo point' , 'Public name' , 'Subscription status'])
df_clean.head()
```

```python
#Checking for duplicates

df_clean.duplicated()
```

```python
#Dropping duplicates

df_clean.drop_duplicates()
```

```python
#Checking for missing values

df_clean.isnull().sum()
```

```python
#Dropping column with missing values.

df_clean = df_clean.drop(columns = ['Scheduled at'])
df_clean.head()
```

```python
#creating a new date and time column.


df_clean['Datetime'] = pd.to_datetime(df_clean[['year','month','day','hour','minute']])

df_clean.head(5)
```

```python
#sorting the data in descending values according to the date

df_clean = df_clean.sort_values(by = 'Datetime')
df_clean.head(5)
```

```python
#Stripping ('.') from the Utilib 1.4 column name.

df.columns = df.columns.str.replace('.', '').str.replace('©','')
df.columns

```

# **DATA ANALYSIS**

```python
# Finding the usage of the blue car by the difference.

df_clean['Utilib_Usage'] = df_clean['Utilib counter'].diff()
df_clean

df_clean['Utilib 14 Usage'] = df_clean['Utilib 1.4 counter'].diff()
df_clean

df_clean['BC_Usage'] = df_clean['Bluecar counter'].diff()
df_clean
```

```python
# Sum of all car usages.

df_clean['sum_column'] = df_clean["BC_Usage"] + df_clean["Utilib_Usage"] + df_clean["Utilib 14 Usage"]

df_clean
```

```python
#Filtering out the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

df_clean1 = df_clean.loc[(df_clean['City'] == 'Paris') & (df_clean['Rental status'] == 'operational') & (df_clean['BC_Usage'] <0)]
df_clean1
```

**1. Most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.**

```python
#single out the time with most picked up cars.


df_clean2 = df_clean1.groupby(['hour'])['BC_Usage'].agg(sum).sort_values(ascending=True)
df_clean2.head(1)
```

```python
#What is the most popular hour for returning cars?
df_clean1 = df_clean.loc[(df_clean['City'] == 'Paris') & (df_clean['Rental status'] == 'operational') & (df_clean['BC_Usage'] >0)]
df_clean1

df_clean3 = df_clean1.groupby(['hour'])['sum_column'].sum().sort_values(ascending=True)
df_clean3.head()
```
